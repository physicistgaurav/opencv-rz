```python
img_mia = cv2.imread("mia.jpg", cv2.IMREAD_COLOR)
img_mia= img_mia[:,:,::-1]
plt.imshow(img_mia)
```

```python
crop = img_mia[700:950, 500:900]
plt.imshow(crop)
```

```python
resize_crop = cv2.resize(crop, None, fx=2, fy=2)
plt.imshow(resize_crop)
```

```python
desired_width = 100
desired_height =200
dim = (desired_width, desired_height)

#resize
resized_cropped_region = cv2.resize(crop, dsize = dim, interpolation = cv2.INTER_AREA)
plt.imshow(resized_cropped_region)
```

```python
#method
desire_width = 50
aspect_ratio = desire_width / crop.shape[1]
desired_height = int(crop.shape[0]*aspect_ratio)
dim = (desire_width, desired_height)

#resize
resized_cropped_region = cv2.resize(crop, dsize= dim, interpolation=cv2.INTER_AREA)
plt.imshow(resized_cropped_region)
```

```python
img_mia_rgb = Image('mia.jpg')
img_mia_flipped_horz = cv2.flip(img_mia, 1)
img_mia_flipped_vert = cv2.flip(img_mia,0)
img_mia_flipped_both = cv2.flip(img_mia,-1)

#show
plt.figure(figsize=[18,5])
plt.subplot(141);plt.imshow(img_mia_flipped_horz);plt.title("horizontal flip");
plt.subplot(142);plt.imshow(img_mia_flipped_vert);plt.title("Vertical flip");
plt.subplot(143);plt.imshow(img_mia_flipped_both);plt.title("Both flip");
plt.subplot(144);plt.imshow(img_mia);plt.title("orgi");
```
